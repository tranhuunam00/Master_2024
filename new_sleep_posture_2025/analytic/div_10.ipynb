```python
import pandas as pd

data = pd.read_csv('../datas/test.csv')

data[['x', 'y', 'z']] = (data[['x', 'y', 'z']] / 10).round(3)

data.to_csv('../datas/test_div10.csv', index=False)

print("✅ Đã tạo file test_div10.csv với x, y, z chia 10 và làm tròn 2 chữ số.")
```
