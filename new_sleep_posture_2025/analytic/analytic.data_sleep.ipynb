```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Đọc dữ liệu
data = pd.read_csv('../datas/train_div10.csv')
data_test = pd.read_csv('../datas/test_div10.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Map số hoạt động thành nhãn dễ hiểu
activity_map = {1: 'Ngửa', 2: 'Nghiêng phải', 3: 'Nghiêng trái', 4: 'Sấp'}
dataDraw = data.copy()

dataDraw['activity'] = dataDraw['activity'].map(activity_map)

# Cấu hình style cho biểu đồ
sns.set_theme(style="whitegrid", font_scale=1.2)
custom_palette = {
    'Ngửa': '#1f77b4',       # Xanh dương
    'Nghiêng phải': '#ff7f0e',       # Cam
    'Nghiêng trái': '#2ca02c',      # Xanh lá
    'Sấp': '#d62728'     # Đỏ
}

# Tạo Figure tổng hợp gồm 3 hàng
fig, axes = plt.subplots(3, 1, figsize=(8, 10), sharex=True)

for i, axis in enumerate(['x', 'y', 'z']):
    sns.boxplot(
        data=dataDraw, 
        x='activity', 
        y=axis, 
        ax=axes[i],
        palette=custom_palette,
        linewidth=1.5,
        fliersize=4,   # kích thước điểm ngoại lai
        width=0.6
    )
    axes[i].set_title(f"Phân bố giá trị trục {axis.upper()} tập huấn luyện", fontsize=14, fontweight='bold')
    axes[i].set_xlabel("Tư thế ngủ", fontsize=12)
    axes[i].set_ylabel(f"Giá trị trục {axis.upper()}", fontsize=12)
    axes[i].grid(True, linestyle='--', alpha=0.5)


plt.tight_layout()
plt.show()
```

```python
activity_map = {1: 'Ngửa', 2: 'Nghiêng phải', 3: 'Nghiêng trái', 4: 'Sấp'}
dataDrawBalance = data.copy()
dataDrawBalance['activity'] = dataDrawBalance['activity'].map(activity_map)

# Đếm số lượng mẫu cho từng tư thế
counts = dataDrawBalance['activity'].value_counts().reset_index()
counts.columns = ['Tư thế ngủ', 'Số lượng mẫu']

# Cấu hình style và màu sắc
sns.set_theme(style="whitegrid", font_scale=1.2)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

# Vẽ biểu đồ cột
plt.figure(figsize=(7, 5))
sns.barplot(
    data=counts,
    x='Tư thế ngủ',
    y='Số lượng mẫu',
    palette=colors,
    edgecolor='black'
)

# Thêm nhãn số lượng trên từng cột
for i, v in enumerate(counts['Số lượng mẫu']):
    plt.text(i, v + (v * 0.01), f"{v:,}", ha='center', va='bottom', fontsize=11, fontweight='bold')

# Thiết lập tiêu đề và nhãn
plt.title("Phân bố số lượng mẫu theo tư thế ngủ tập huấn luyện", fontsize=15, fontweight='bold')
plt.xlabel("Tư thế ngủ", fontsize=12)
plt.ylabel("Số lượng mẫu", fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Map lại tên và giữ thứ tự 1-2-3-4
activity_order = ['Ngửa', 'Nghiêng phải', 'Nghiêng trái', 'Sấp']
activity_map = {1: 'Ngửa', 2: 'Nghiêng phải', 3: 'Nghiêng trái', 4: 'Sấp'}

# Tạo bản sao dữ liệu để không thay đổi gốc
dataDesc = data.copy()
dataDesc['activity'] = dataDesc['activity'].map(activity_map)

# Tính thống kê mô tả
desc = dataDesc.groupby('activity', observed=False)[['x', 'y', 'z']].describe().round(3)
print("=== THỐNG KÊ MÔ TẢ THEO TƯ THẾ NGỦ ===")
print(desc)

# Trích riêng trung bình và độ lệch chuẩn
mean_values = desc.xs('mean', level=1, axis=1).reindex(activity_order)
std_values = desc.xs('std', level=1, axis=1).reindex(activity_order)

# Cấu hình hiển thị
sns.set_theme(style="whitegrid", font_scale=1.3)
plt.figure(figsize=(9,6))

# Vẽ biểu đồ trung bình + thanh sai số
mean_values.plot(
    kind='bar',
    yerr=std_values,
    capsize=5,
    figsize=(9,6),
    color=['#1f77b4','#ff7f0e','#2ca02c'],
    edgecolor='black',
    width=0.75,
    alpha=0.9
)

plt.title('Giá trị trung bình và độ lệch chuẩn của từng trục theo tư thế ngủ',
          fontsize=15, fontweight='bold')
plt.xlabel('Tư thế ngủ', fontsize=13)
plt.ylabel('Giá trị trung bình (± độ lệch chuẩn)', fontsize=13)
plt.xticks(rotation=0)
plt.legend(title='Trục cảm biến', fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

for act, color in custom_palette.items():
    subset = dataDraw[dataDraw['activity'] == act]
    ax.scatter(subset['x'], subset['y'], subset['z'], label=act, alpha=0.5, color=color, s=10)

ax.set_xlabel('Trục X')
ax.set_ylabel('Trục Y')
ax.set_zlabel('Trục Z')
ax.legend()
plt.title('Phân bố không gian 3D của dữ liệu gia tốc', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

```
