```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Đọc dữ liệu
data = pd.read_csv('totalData2.csv')
data_test = pd.read_csv('new_nam.csv')

# Rút gọn timestamp
data['createdAt'] = data['createdAt'].str[:-38]
data_test['createdAt'] = data_test['createdAt'].str[:-38]

# Chuyển activity từ số sang tên
activity_labels = {
    1: 'Back',
    2: 'Left',
    3: 'Right',
    4: 'Stomach'
}
data['activity'] = data['activity'].astype(int).replace(activity_labels)

# Thiết lập style và font
sns.set(style="whitegrid")
plt.rc('font', weight='bold', size=13)

# Tạo bảng màu tươi cho từng activity
activities = data['activity'].unique()
bright_colors = sns.color_palette("bright", len(activities))
activity_palette = dict(zip(activities, bright_colors))

# Tạo figure nhỏ gọn
fig, axes = plt.subplots(3, 1, figsize=(4.5, 6.5))  # Có thể dùng (5, 7.5) nếu muốn lớn hơn

# Boxplot trục x
sns.boxplot(data=data, x='activity', y='x', hue='activity', ax=axes[0],
            palette=activity_palette, legend=False)
axes[0].set_title('Distribution of x-axis values', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Activity', fontsize=11, fontweight='bold')
axes[0].set_ylabel('x', fontsize=11, fontweight='bold')
axes[0].tick_params(axis='both', labelsize=10)

# Boxplot trục y
sns.boxplot(data=data, x='activity', y='y', hue='activity', ax=axes[1],
            palette=activity_palette, legend=False)
axes[1].set_title('Distribution of y-axis values', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Activity', fontsize=11, fontweight='bold')
axes[1].set_ylabel('y', fontsize=11, fontweight='bold')
axes[1].tick_params(axis='both', labelsize=10)

# Boxplot trục z
sns.boxplot(data=data, x='activity', y='z', hue='activity', ax=axes[2],
            palette=activity_palette, legend=False)
axes[2].set_title('Distribution of z-axis values', fontsize=12, fontweight='bold')
axes[2].set_xlabel('Activity', fontsize=11, fontweight='bold')
axes[2].set_ylabel('z', fontsize=11, fontweight='bold')
axes[2].tick_params(axis='both', labelsize=10)

# Căn chỉnh layout và hiển thị
plt.tight_layout()
plt.show()
```
